```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_excel('DataScientist_009749_Dataset.xlsx', index_col=0)
```

```python
test.head()
```

```python
test.columns = pd.MultiIndex.from_arrays([test.columns, test.iloc[0].values])
test = test.iloc[1:]
```

```python
test
```

```python
df1 = pd.melt(test, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=False)
```

```python
var_list = ["NWP (£m) ","SCR (£m)","Excess of assets over liabilities (£m) [= equity]","EoF for SCR (£m)","Total assets (£m)","SCR coverage ratio","GWP (£m)","Total liabilities (£m)"]
```

```python
num_list = ["1","2","3","4"]
```

```python
for i in var_list:
    for j in num_list:
        rep = i+"."+j
        df1["variable_0"] = np.where(df1["variable_0"]==rep,i,df1["variable_0"])
```

```python
df1["variable_0"].value_counts()
```

```python
df1
```

```python
df1 = df1.reset_index()
```

```python
df1 = df1.pivot(index = ["index","variable_1"],columns='variable_0', values='value')
```

```python
df1 = df1.reset_index()
```

```python
df1.columns
```

```python
test1 = pd.read_excel('DataScientist_009749_Dataset.xlsx', sheet_name="Dataset 2 - Underwriting", index_col=0)
```

```python
test1
```

```python
test1.columns = pd.MultiIndex.from_arrays([test1.columns, test1.iloc[0].values])
test1 = test1.iloc[1:]
```

```python
test1
```

```python
df2 = pd.melt(test1, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=False)
```

```python
var_list = ["Net expense ratio","Pure gross claims ratio","Gross expense ratio","Gross combined ratio",
            "Net combined ratio","Gross claims incurred (£m)","Gross BEL (inc. TPs as whole, pre-TMTP) (£m)",
            "Net BEL (inc. TPs as a whole, pre-TMTP) (£m)","Pure net claims ratio"]
```

```python
num_list = ["1","2","3","4"]
```

```python
for i in var_list:
    for j in num_list:
        rep = i+"."+j
        df2["variable_0"] = np.where(df2["variable_0"]==rep,i,df2["variable_0"])
```

```python
df2["variable_0"].value_counts()
```

```python
df2
```

```python
df2 = df2.reset_index()
```

```python
df2 = df2.pivot(index = ["index","variable_1"],columns='variable_0', values='value')
```

```python
df2 = df2.reset_index()
```

```python
df2.dtypes
```

```python
df2.describe()
```

```python
df1
```

```python
merged = df1.merge(df2,how="inner",on=["index","variable_1"])
```

```python
merged = merged.sort_values(by=['company', 'year'])
```

```python
merged
```

```python
#convert to numerical
```

```python
merged = merged.rename(columns = {"index":"company","variable_1":"year"})
```

```python
merged['company'] = merged['company'].str[4:]
```

```python
merged['year'] = merged['year'].str[:-2]
```

```python
merged.dtypes
```

```python
#check for missing values
merged.isna().sum()
```

```python
merged = merged.apply(pd.to_numeric)
```

```python
merged.to_csv("data.csv")
```
